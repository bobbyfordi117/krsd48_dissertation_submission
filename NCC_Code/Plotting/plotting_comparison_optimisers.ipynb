```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#set font size to 16
plt.rcParams.update({'font.size': 16})

#colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

df_adam_accuracy = pd.read_csv('Optimiser/optimiser_accuracy/CRAS-103__train_epoch_accuracy.csv')
df_adamax_accuracy = pd.read_csv('Optimiser/optimiser_accuracy/CRAS-104__train_epoch_accuracy.csv')
df_rms_accuracy = pd.read_csv('Optimiser/optimiser_accuracy/CRAS-106__train_epoch_accuracy.csv')
df_sgd_accuracy = pd.read_csv('Optimiser/optimiser_accuracy/CRAS-101__train_epoch_accuracy.csv')


df_adam_loss = pd.read_csv('Optimiser/optimiser_loss/CRAS-103__train_epoch_loss.csv')
df_adamax_loss = pd.read_csv('Optimiser/optimiser_loss/CRAS-104__train_epoch_loss.csv')
df_rms_loss = pd.read_csv('Optimiser/optimiser_loss/CRAS-106__train_epoch_loss.csv')
df_sgd_loss = pd.read_csv('Optimiser/optimiser_loss/CRAS-101__train_epoch_loss.csv')

df_adam_inference = pd.read_csv('Optimiser/optimiser_inference/CRAS-103__train_epoch_inference.csv')
df_adamax_inference = pd.read_csv('Optimiser/optimiser_inference/CRAS-104__train_epoch_inference.csv')
df_rms_inference = pd.read_csv('Optimiser/optimiser_inference/CRAS-106__train_epoch_inference.csv')
df_sgd_inference = pd.read_csv('Optimiser/optimiser_inference/CRAS-101__train_epoch_inference.csv')

epochs = df_adam_accuracy.iloc[:,0]

accuracy_adam = df_adam_accuracy.iloc[:,2]
accuracy_adamax = df_adamax_accuracy.iloc[:,2]
accuracy_rms = df_rms_accuracy.iloc[:,2]
accuracy_sgd = df_sgd_accuracy.iloc[:,2]

loss_adam = df_adam_loss.iloc[:,2]
loss_adamax = df_adamax_loss.iloc[:,2]
loss_rms = df_rms_loss.iloc[:,2]
loss_sgd = df_sgd_loss.iloc[:,2]

inference_adam = df_adam_inference.iloc[:,2]
inference_adamax = df_adamax_inference.iloc[:,2]
inference_rms = df_rms_inference.iloc[:,2]
inference_sgd = df_sgd_inference.iloc[:,2]

#colourmap for four optimisers
cmap = plt.get_cmap('Set2')
```

```python
plt.figure()
plt.plot(epochs, accuracy_adam, label='Adam', color=cmap(0))
plt.plot(epochs, accuracy_adamax, label='Adamax', color=cmap(1))
plt.plot(epochs, accuracy_rms, label='RMS', color=cmap(3))
plt.plot(epochs, accuracy_sgd, label='SGD', color=cmap(2))
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('optimisers_accuracy.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, loss_adam, label='Adam', color=cmap(0))
plt.plot(epochs, loss_adamax, label='Adamax', color = cmap(1))
plt.plot(epochs, loss_rms, label='RMS', color = cmap(3))
plt.plot(epochs, loss_sgd, label='SGD', color = cmap(2))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('optimisers_loss.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, inference_adam, label='Adam', color = cmap(0))
plt.plot(epochs, inference_adamax, label='Adamax', color = cmap(1))
plt.plot(epochs, inference_rms, label='RMS', color = cmap(3))
plt.plot(epochs, inference_sgd, label='SGD', color = cmap(2))
plt.xlabel('Epochs')
plt.ylabel('Inference')
plt.legend()
plt.savefig('optimisers_inference.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```
