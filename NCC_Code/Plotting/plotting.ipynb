```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv('train_epoch_accuracy.csv')
df2 = pd.read_csv('train_epoch_loss.csv')

epochs = df1.iloc[:,0]
accuracy = df1.iloc[:,2]
loss = df2.iloc[:,2]

plt.figure()
plt.plot(epochs, accuracy, label='Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('train_accuracy.png')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, loss, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('train_loss.png')
plt.show()
```
