```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 24})

df_mel_accuracy = pd.read_csv('Mel vs Spec/mel__train_epoch_accuracy.csv')
df_norm_accuracy = pd.read_csv('Mel vs Spec/norm__train_epoch_accuracy.csv')

df_mel_inference = pd.read_csv('Mel vs Spec/mel__train_epoch_inference.csv')
df_norm_inference = pd.read_csv('Mel vs Spec/norm__train_epoch_inference.csv')

df_mel_loss = pd.read_csv('Mel vs Spec/mel__train_epoch_loss.csv')
df_norm_loss = pd.read_csv('Mel vs Spec/norm__train_epoch_loss.csv')

epochs = df_mel_accuracy.iloc[:,0]

accuracy_mel = df_mel_accuracy.iloc[:,2]
accuracy_norm = df_norm_accuracy.iloc[:,2]

inference_mel = df_mel_inference.iloc[:,2]
inference_norm = df_norm_inference.iloc[:,2]

loss_mel = df_mel_loss.iloc[:,2]
loss_norm = df_norm_loss.iloc[:,2]
```

```python
plt.figure()
plt.plot(epochs, accuracy_mel, label='Mel')
plt.plot(epochs, accuracy_norm, label='Norm')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('melVnorm_accuracy.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, loss_mel, label='Mel')
plt.plot(epochs, loss_norm, label='Norm')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('melVnorm_loss.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, inference_mel, label='Mel')
plt.plot(epochs, inference_norm, label='Norm')
plt.xlabel('Epochs')
plt.ylabel('Inference')
plt.legend()
plt.savefig('melVnorm_inference.png' ,dpi = 300 , bbox_inches='tight')
plt.show()
```
