```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#set font size to 16
plt.rcParams.update({'font.size': 16})

#color palette
colors = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00','#ffff33','#a65628','#f781bf','#999999']

df_aug_accuracy = pd.read_csv('Grades Noise Aug & Std/aug__train_epoch_accuracy.csv')
df_std_accuracy = pd.read_csv('Grades Noise Aug & Std/std__train_epoch_accuracy.csv')
df_noise_accuracy = pd.read_csv('Grades Noise Aug & Std/noise__train_epoch_accuracy.csv')

df_aug_loss = pd.read_csv('Grades Noise Aug & Std/aug__train_epoch_loss.csv')
df_std_loss = pd.read_csv('Grades Noise Aug & Std/std__train_epoch_loss.csv')
df_noise_loss = pd.read_csv('Grades Noise Aug & Std/noise__train_epoch_loss.csv')

df_aug_inference = pd.read_csv('Grades Noise Aug & Std/aug__train_epoch_inference.csv')
df_std_inference = pd.read_csv('Grades Noise Aug & Std/std__train_epoch_inference.csv')
df_noise_inference = pd.read_csv('Grades Noise Aug & Std/noise__train_epoch_inference.csv')

epochs = df_aug_accuracy.iloc[:,0]

accuracy_aug = df_aug_accuracy.iloc[:,2]
accuracy_std = df_std_accuracy.iloc[:,2]
accuracy_noise = df_noise_accuracy.iloc[:,2]

loss_aug = df_aug_loss.iloc[:,2]
loss_std = df_std_loss.iloc[:,2]
loss_noise = df_noise_loss.iloc[:,2]

inference_aug = df_aug_inference.iloc[:,2]
inference_std = df_std_inference.iloc[:,2]
inference_noise = df_noise_inference.iloc[:,2]
```

```python
plt.figure()
plt.plot(epochs, accuracy_aug, label='Original', color = colors[0])
plt.plot(epochs, accuracy_std, label='Standard', color = 'dimgrey')
plt.plot(epochs, accuracy_noise, label='ANF', color = 'purple')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('grades_accuracy.png',dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, loss_aug, label='Original', color = colors[0])
plt.plot(epochs, loss_std, label='Standard', color = 'dimgrey')
plt.plot(epochs, loss_noise, label='ANF', color = 'purple')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('grades_loss.png',dpi = 300 , bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.plot(epochs, inference_aug, label='Original', color = colors[0])
plt.plot(epochs, inference_std, label='Standard', color = 'dimgrey')
plt.plot(epochs, inference_noise, label='ANF', color = 'purple')
plt.xlabel('Epochs')
plt.ylabel('Inference')
plt.legend()
plt.savefig('grades_inference.png',dpi = 300 , bbox_inches='tight')
plt.show()
```
